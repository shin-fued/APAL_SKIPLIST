```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cap_change.csv')
df["log n"] = np.log(df["n"])
df["time sk log"]=np.log(df[" time skip_list  ns"])
df["time map log"]=np.log(df["time map  ns"])
df.columns
```

```python
x1 = np.linspace(8, 18, 10000)
```

```python
ins_df = df[df[' operation'] == 'insert']
sea_df = df[df[' operation'] == 'search']
del_df = df[df[' operation'] == 'delete']
plt.plot(ins_df['log n'], ins_df['time sk log'], 'x', label='insert skip')
plt.plot(ins_df['log n'], ins_df['time map log'], 'o', label='insert map')

plt.xlabel("log size during insert")
plt.ylabel("log time taken ns")
plt.legend()
```

```python
plt.plot(sea_df['log n'], sea_df['time sk log'], 'x',label='search skip')
plt.plot(sea_df['log n'], sea_df['time map log'], 'o',label='search map')
plt.xlabel("log size during search")
plt.ylabel("log time taken ns")

plt.legend()
```

```python
plt.plot(del_df['log n'], del_df['time sk log'], 'x', label='delete skip')
plt.plot(del_df['log n'], del_df['time map log'], 'o', label='delete map')
plt.xlabel("log size deleted")
plt.ylabel("log time taken ns")
plt.legend()
```

```python
data = pd.read_csv('scale_time.csv')
data["log n"] = np.log(data["n"])
data["time sk log"]=np.log(data[" time skip_list ns"])
data["time map log"]=np.log(data["time map ns"])
ins_d = data[data[' operation'] == 'insert']
sea_d = data[data[' operation'] == 'search']
del_d = data[data[' operation'] == 'delete']
```

```python
ins_skip = ins_d.groupby('log n')['time sk log'].min()
ins_map = ins_d.groupby('log n')['time map log'].min()
plt.plot(ins_skip.index, ins_skip.values, label='insert skip')
plt.plot(ins_map, ins_map.values, label='insert map')
plt.xlabel("log of insert size")
plt.ylabel("log min time taken ns")
plt.legend()
```

```python
sea_skip = sea_d.groupby('log n')['time sk log'].min()
sea_map = sea_d.groupby('log n')['time map log'].min()
plt.plot(sea_skip.index, sea_skip.values, label='search skip')
plt.plot(sea_map, sea_map.values, label='search map')
plt.xlabel("log of search size")
plt.ylabel("log min time taken ns")
plt.legend()
```

```python
del_skip = del_d.groupby('log n')['time sk log'].min()
del_map = del_d.groupby('log n')['time map log'].min()
plt.plot(del_skip.index, del_skip.values, label='delete skip')
plt.plot(del_map, del_map.values, label='delete map')
plt.xlabel("log of delete size")
plt.ylabel("log min time taken ns")
plt.legend()
```
